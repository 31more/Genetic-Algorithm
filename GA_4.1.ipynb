```python
import numpy as np
import pandas as pd
import pygad
```

```python
file = pd.read_csv('11.txt', delimiter = ' ')
file
```

```python
params = file.columns.tolist()
W = int(params[0])
V = int(params[1])
file.columns = ['v', 'c']
file['w'] = file.index
data = file[['w','v','c']].values.tolist()
data
```

```python
def fitness(solution, solution_index):
    w = 0 
    v = 0
    c = 0 
    for (array, item) in zip(solution, data):
        if array > 0:
            w += item[0]
            v += item[1]
            c += item[2]
    if w > W or v > V:
        c = 0
    return c
```

```python
ga_instance = pygad.GA(num_generations=250, #Количество поколений
                       num_parents_mating=2, #Количество решений, которые будут выбраны в качестве родителей
                       fitness_func=fitness, #Функция, вычисляющая значение пригодности решения
                       sol_per_pop=35, #Количество решений в популяции
                       num_genes=len(data), #Количество генов в решении
                       init_range_low=-1, #Нижнее значение случайного диапазона, из которого выбираются значения генов в исходной популяции
                       init_range_high=1 #Верхнее значение случайного диапазона, из которого выбираются значения генов в исходной популяции
                       )
```

```python
ga_instance.run()
solution, solution_fitness, solution_idx = ga_instance.best_solution()
```

```python
result = []
for i in range(0, len(data)):
    if solution[i] > 0:
        result.append(data[i])
result_file = pd.DataFrame.from_records(result, columns=['w','v','c'])
```

```python
print(solution_fitness)
print(result_file)
w = 0
v = 0
c = 0
for i in range(0, len(result_file)):
        w += result_file.at[i,'w']
        v += result_file.at[i,'v']
        c += result_file.at[i,'c']
print("Вес:", w, "(Грузоподъемноcть:", W, ")")
print("Объем:", round(v,1), "(Вместимость:", V, ")")
print("Ценность:", c)
```

```python
file=open('4.1.txt','w', encoding='utf-8')
file.write(result_file.to_string())
file.write("\nВес: " + str(w) + " (Грузоподъемноcть: " + str(W) + ")")
file.write("\nОбъем: " + str(round(v,1)) + " (Вместимость: " + str(V) + ")")
file.write("\nЦенность: " + str(c))
file.close()
```
